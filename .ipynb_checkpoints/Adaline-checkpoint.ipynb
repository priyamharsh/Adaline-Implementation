```python
import numpy as np
```

**Defining the Sign Function which is activation function**

```python
def sign_func(x, theta=0):
    if x>theta:
        return 1
    elif x>=(-1*theta) and x<=theta:
        return 0
    elif x<(-1*theta):
        return -1
```

## Implementing the Adaline Class

```python
class Adaline():
    def __init__(self):
        self.weights = []
    
    def train_NN(self, X, Y, alpha=0.1, epoch=10):
        # initializing weights
        weights = np.array([0]+[0 for i in range(len(X[0]))])
        # initializing bias
        b = np.array([[1 for i in range(len(X))]])
        X = np.concatenate((X, b.T), axis=1)
        epsilon = 0.01
        count = 0
        stoppin_criterion = False
        while stoppin_criterion==False:
            mse = np.array([])
            for i in range(len(X)):
                # calculating net input
                yin = sum(X[i]*weights)
                del_weights = np.array([])
                for j in range(len(X[i])):
                    # calculating change in weights
                    del_weights = np.append(del_weights, [X[i, j]*(Y[i]-yin)*alpha])
                weights = weights + del_weights
                mse = np.append(mse, 0.5*((Y[i]-yin))**2)
            count += 1
            if count>=epoch or sum(mse)<=epsilon:
                stoppin_criterion = True
        self.weights = weights
    
    def get_weights(self):
        return self.weights
    
    def predict(self, X):
        X_in = np.array([i.flatten() for i in X])
        b = np.array([[1 for i in range(len(X))]])
        X_w_bias = np.concatenate((X_in, b.T), axis=1)
        predictions = [sum(X_w_bias[i]*self.weights) for i in range(len(X))]
        return predictions
```

## Creating the time series data

```python
t1 = np.arange(0, 3.05, 0.01)
t2 = np.arange(3.05, 6.10, 0.01)
t0 = np.concatenate((t1, t2))
```

```python
t0.shape
```

```python
y = np.concatenate((np.sin(4.1*np.pi*t1), 0.8*np.sin(8.3*np.pi*t2)))
```

Creating some data with Gaussian noise

```python
noise = np.random.normal(0,1,610)
y_noise = y + 0.1*noise
```

**Plotting the data**

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 10))
plt.plot(t0, y)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(t0, y_noise)
```

### Creating data with previous values to simulate delay

```python
X = []       #past 5 values
t = []       #current target
X_noise = [] #past 5 values
t_noise = [] #current target
for i in range(5, len(y)):
  X.append(y[i-5:i])
  t.append(y[i])
  X_noise.append(y_noise[i-5:i])
  t_noise.append(y_noise[i])
```

```python
X = np.array(X)
t = np.array(t)
X_noise = np.array(X_noise)
t_noise = np.array(t_noise)
```

## Training model and plotting results

```python
model1 = Adaline()
model1.train_NN(X, t, alpha=0.01, epoch=500)
ypred_2 = model1.predict(X)
```

```python
ypred_2 = np.array(ypred_2)
```

```python
plt.figure(figsize=(10, 10))
plt.plot(t0[5:].reshape(-1), t.reshape(-1))
plt.plot(t0[5:].reshape(-1), ypred_2.reshape(-1))
plt.legend(['Original','Predicted'])
```

```python
model2 = Adaline()
model2.train_NN(X_noise, t_noise, alpha=0.01, epoch=500)
y2 = model1.predict(X_noise)
y2 = np.array(y2)
plt.figure(figsize=(10, 10))
plt.plot(t0[5:].reshape(-1), t_noise.reshape(-1))
plt.plot(t0[5:].reshape(-1), y2.reshape(-1))
plt.legend(['Original','Predicted'])
```

# Conclusion


*   We can see from the results that a simple Adaline neuron can model complex time series plots with minimum error. Even when Gaussian noise is added to the data, the model predicts the correct trend most of the time.
*   Some error can be observed when the frequency and the amplitude of the wave changes. This is because, the model cannot predict correct values based on previous data that has different amplitude and frequncy.
